```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from benchmark import run_experiments
import os
import os.path

here = os.path.dirname(os.path.dirname("__file__"))
PROJECT_DIR = os.path.abspath(os.path.join(here, "../../.."))

PATHS = {
  "DNA" : "exp/crossings/dna/results/rm-versions",
  "morphemes": "exp/crossings/morphemes/2grams/results/rm-versions"
}

CHOSEN_DS = "DNA"

RESULTS_DIR = os.path.join(PROJECT_DIR, PATHS[CHOSEN_DS])
```

```python
T = pd.read_csv(os.path.join(RESULTS_DIR, "time.csv"))
M = pd.read_csv(os.path.join(RESULTS_DIR, "mem.csv"))
O = pd.read_csv(os.path.join(RESULTS_DIR, "outputs.csv"))
F = pd.read_csv(os.path.join(RESULTS_DIR, "foutt.csv"))
```

```python
sers = {
"Time": {
  "data": T[[f"REm-{i}" for i in range(0,6)]],
  "units": "ms"
  },
"Outputs":  {
  "data": O[[f"REm-{i}" for i in range(0,6)]],
  "units": ""
  },
"Memory":   {
  "data": M[[f"REm-{i}" for i in range(1,5)]],
  "units": "kB"
  },
"FOT": {
  "data": F[[f"REm-{i}" for i in range(0,6)]],
  "units": "ms"
  }
}
```

```python
for (k, v) in sers.items():
  fig_mem = plt.figure()
  bp = v["data"].boxplot()
  plt.title(f"{CHOSEN_DS} sample: {k}")
  plt.ylabel(v["units"])
  fig_mem.savefig(os.path.join(RESULTS_DIR, f'{k[0].lower()}-bp.png'))
```
